### Compute 2D FFT of the images

#### My FFT

```python
def separate(x):
    return x[::2], x[1::2]

def MyFFT(x, N):
    if N < 2: return x 
    x_even, x_odd = separate(x)
    X1 = MyFFT(x_even, N//2)
    X2 = MyFFT(x_odd, N//2)
    W_K = [np.exp(-2j*np.pi*k/N)*X2[k] for k in range(len(X2))]
    return ([X1[i] + W_K[i] for i in range(len(X1))] + [X1[i] - W_K[i] for i in range(len(X1))])
```

#### My FFT 2D

```python
def My2DFFT(x):
    output = np.zeros(x.shape, dtype = 'complex_')
    #Iterate on s and compute 1D FFTs
    for s in range(x.shape[0]):
        output[s, :] = MyFFT(x[s, :], len(x))

    #Iterate on t and compute 1D FFTs
    for t in range(x.shape[1]):
        output[:, t] = MyFFT(output[:, t], len(x))
    return output
```

#### A - FFT of image a

```python
A = My2DFFT(a)
A_magnitude = np.absolute(A)
A_phase = np.angle(A)
A_shifted = np.fft.fftshift(A)
plt.imshow(np.log10(1 + np.absolute(A_shifted)))
plt.title("Zero-centered 2D FFT: Image 'a'")
```

#### B - FFT of image b

```python
B = My2DFFT(b)
B_magnitude = np.absolute(B)
B_phase = np.angle(B)
B_shifted = np.fft.fftshift(B)
plt.imshow(np.log10(1 + np.absolute(B_shifted)))
plt.title("Zero-centered 2D FFT: Image 'b'")
```

### Swap phases and generate images 

#### Inverse FFT

```python
def MyIFFT(x):
    N = x.shape[0]

    #Complex conjugate of x
    x_conjugate = np.conjugate(x)

    #Compute fft of the conjugate
    x_fft = MyFFT(x_conjugate, len(x_conjugate))

    #Complex conjugate of the fft of x
    x_fft_conjugate = np.conjugate(x_fft)
    
    #Divide by N
    result = x_fft_conjugate / N
    return result
```

#### Inverse FFT 2D

```python
def My2DIFFT(x):
    output = np.zeros(x.shape, dtype = 'complex_')
    for s in range(x.shape[0]):
        output[s, :] = MyIFFT(x[s, :])

    for t in range(x.shape[1]):
        output[:, t] = MyIFFT(output[:, t])
    return output.real
```
