## III. Spatial Transformations

For the following, the output image should be the same size as the input image. Outside of boundary values can default to zero. You are allowed to use a readily available interpolation function. Note that the “transformed image” might not have the field view to show the entire original scene.

```python
x = img.imread('./X.png')
plt.imshow(x, cmap="gray")
plt.title('Original image')
```

#### Rotate 2D

Write a function called Rotate2D that accepts an input image and a rotation angle (in degrees) and outputs an image that is the input rotated around the center pixel of the image.


```python
def Rotate2D(image, angle):
    #Dimensions
    height = image.shape[0]
    width = image.shape[1]
    
    # Find the origin of the image
    y_center = round(((height + 1)/2)-1)  
    x_center = round(((width + 1)/2)-1)
    
    # Output matrix with the same size as the input image
    rotated_image = np.zeros(image.shape)

    #Iterate every pixel
    for h in range(height):
        for w in range(width):
            
            ##Intensity value
            value = image[h,w]
            
            #location relative to the center
            y = height-1-h-y_center                   
            x = width-1-w-x_center                      

            #apply rotation transform to find new values
            y_new = round(-x * np.sin(angle) + y * np.cos(angle))
            x_new = round(x * np.cos(angle) + y * np.sin(angle))

            #location relative to origin
            y_new = y_center - y_new
            x_new = x_center - x_new

            #Append values if the location is within the valid frame
            if y_new < height and x_new < width: 
                #Assign pixel value to the new locations
                rotated_image[y_new,x_new] = value 
    return rotated_image
```

Read the image X.png. Rotate it by 45 degrees, save this as RotatedX.png, and visualize this image.

```python
# Rotate image
x_rotated = Rotate2D(x, 45)

# Save image
im = Image.fromarray((255 * x_rotated).astype(np.uint8))
im.save('RotatedX.png')

# Read and display image
x_rotated_image = img.imread('./RotatedX.png')
plt.imshow(x_rotated_image, cmap="gray")
plt.title('Image rotated 45°')
```

#### Translate 2D

Write a function called Translate2D that accepts an input image and a translation (shift, displacement) vector (in 2D, pixel units); and outputs an image that is the input image shifted/translated by the provided displacement vector.

```python
def Translate2D(image, shift):
    height = image.shape[0]
    width = image.shape[1]
    y_shift = shift[0]
    x_shift = shift[1]

    # Output matrix with the same size as the input image
    translated_image = np.zeros(image.shape)

    #Iterate every pixel
    for y in range(height):
        for x in range(width):
            
            ##Intensity value
            value = image[y,x]
            
            #Apply shift
            y_new = y + y_shift
            x_new = x + x_shift
            
            #If new position exceeds size, break to return black pixel
            if y_new >= height or x_new >= width: break

            #Handle negative shifts
            if (y_shift < 0 and x_new >= 0 and y_new >= 0) or (y_shift > 0 and x_new >= 0): 
                translated_image[y_new, x_new] = value
                
    return translated_image
```

Shift X by [30, −70] pixels, save this as ShiftedX.png, and visualize this image.

```python
# Shift image
shift = np.array([30, -70])
x_shifted = Translate2D(x, shift)

# Save image
im = Image.fromarray((255 * x_shifted).astype(np.uint8))
im.save('ShiftedX.png')

# Read and show image
x_shifted_image = img.imread('./ShiftedX.png')
plt.imshow(x_shifted_image, cmap="gray")
plt.title('Image shifted by [30, -70]')
```
